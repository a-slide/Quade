## Cross contamination of sample in NGS due to wrong attribution of index

We experienced some problems of incorrect read attribution during Illumina read demultiplexing using simple sample indexing with CASAVA. This led to a cross contamination of reads from one sample to another, which can be problematic for low frequency variant detection or other specific usage of NGS. An experimental approach to reduce the number of incorrectly attributed cluster is to add a second sample index in p5 adapter. Bioinformatics can also provide an additional answer by taking into account the PHRED quality of index read(s). CASAVA is not able to manage quality of INDEX reads, but by modifying the sample sheet, it's possible to output non demultiplexed fastq files (see [convert-bcl-to-fastq](https://gist.github.com/brantfaircloth/3125885)). In this case, index read(s) will be output in separate fastq file(s) synchronized with read1 and read2. I want to design an Illumina paired end read demultiplexer to attribute each read pair to a given sample based on the quality of the index read.


## Quantificative bias for high depth sequencing.

When the depth of coverage (DOC) is very high, PCR duplicates and independent sonication fragment cannot by differentiate. Given that the sequence composition and secondary structures can influence the efficiency of PCR and sequencing steps, quantifying sequences with heterogeneous DOC is inherently biased. To correct this bias, it is possible to add a stochastic sequence after the sample index (several N). This sequences also called molecular index will be included at the end of INDEX reads and can be subsequently used to differentiate PCR replicates by checking if read mapped at the same position have different molecular index. Using such molecular index required modification of bioinformatic tools, to extract the sequence of the molecular index during demultiplexing.  

**To summarize, the demultiplexer need to handle dual sample indexing, including INDEX read quality, and should be able to extract one or 2 molecular index**.

## Defining program IO

### Input

Without demultiplexing, CASAVA generates 1 file per sequencing reading:
   
* R1 = Forward insert
* R2 = First index
* R3 = Reverse insert
* [R4 = Second index read] (if a second index is used)

Reads in files are synchronized, so that for instance the 4rd read of R1 correspond to the same cluster than the 4rd read of R2, R3 and R4. Files are usually divided in ~ 600 mo chunks.
I found file paths simpler to be indicated as CLI arguments and the other options in a configuration file that could be subsequently parsed by the program.
Files path to fastq files will be required as CLI arguments in a specific order: 

chunk1-R1.fastq  chunk1-R2.fastq  chunk1-R3.fastq chunk1-R4.fastq  chunk2-R1.fastq  chunk2-R2.fastq  chunk2-R3.fastq  chunk2-R4.fastq ...

The configuration file will contain values such as the sequence of index for each sample, the minimal phred quality of index... I usually write first a draft configuration file that can evolved during program dev to include new options.
I know that my configuration file will be parse by the python module ConfigParser, so I wrote the following conf file compatible with its simple syntax. 

### Output

I want to obtain a couple of fastq.gz files for each sample, containing reads corresponding to the sample index specified in the configuration file. The name of each read needs to contain sequences from the sample and molecular index, in order to be easily accessible for post-treatments.  

If index quality is not good enough or if the index (or index combination) does not match any sample, these reads should be stored in unidentified fastq files.

Finally, I also want the program to generate a simple text report summarizing what was done during the analysis.

## Parsing the configuration file

Generally, I like to use python class to handle my main instructions. Configuration file and CLI argument parsing, value verifications and variable storing can be elegantly integrated into the initialization of the class (__init__).

Top level running instruction can then be written in the magic call method to make your object directly callable.

Here is a first version of the main class adapted for interactive interpreters

```python
# Imports
import ConfigParser

# Definition of the main class
class Main(object):
    
    # Class init method 
    def __init__ (self, conf_file): # conf_file arg is only for interactive interpreter. Will be remove in final program
        
        # Define a configuration file parser object
        self.conf = ConfigParser.RawConfigParser(allow_no_value=True)
        
        # Indicate the path to the configuration file
        self.conf.read(conf_file)
        
        # In a try block due to possible errors while reading the file 
        try:
            #### PARSE CONF FILE ####
            
            # Get Values from the required section in the appropriate type 
            self.outdir = self.conf.get("general", "outdir") # for string values
            self.mean_qual = self.conf.getint("quality", "mean_qual") # for integer values
            self.minimal_qual = self.conf.getint("quality", "minimal_qual")
            self.quality_scale = self.conf.get("quality", "quality_scale")
            
            # for each index verify its existence and store start and end positions if needed 
            self.index1 = self.conf.getboolean("index", "index1")  # for boolean values
            if self.index1:
                self.index1_start = self.conf.getint("index", "index1_start")
                self.index1_end = self.conf.getint("index", "index1_end")
            self.index2 = self.conf.getboolean("index", "index2")
            if self.index2:
                self.index2_start = self.conf.getint("index", "index2_start")
                self.index2_end = self.conf.getint("index", "index2_end")
            self.molecular1 = self.conf.getboolean("index", "molecular1")
            if self.molecular1:
                self.molecular1_start = self.conf.getint("index", "molecular1_start")
                self.molecular1_end = self.conf.getint("index", "molecular1_end")
            self.molecular2 = self.conf.getboolean("index", "molecular1")
            if self.molecular2:
                self.molecular2_start = self.conf.getint("index", "molecular2_start")
                self.molecular2_end = self.conf.getint("index", "molecular2_end")
            
            # Sample section are a special case since the number of sample section is variable
            # Only section with a name starting by "sample" will be parsed
            self.sample_list = []
            for sample in [i for i in parser.sections() if i.startswith("sample")]:
                self.sample_list.append({
                    'name'   : self.conf.get(sample, "name"),
                    'index1_seq' :  self.conf.get(sample, "index1_seq"),
                    'index2_seq' :  self.conf.get(sample, "index2_seq")})
            
            #### TEST VALUES ####
            
            # Now all variable can be tested and an assertion error will be raised in case of invalid value
            self.outdir = self.outdir if self.outdir else "./"
            self.quality_scale = self.conf.get("quality", "quality_scale")
            
            assert self.mean_qual >= 0 and self.mean_qual <= 40, "Autorized values for mean_qual : 0 to 40"
            assert self.mean_qual >= 0 and self.mean_qual <= 40, "Autorized values for minimal_qual : 0 to 40"
            if self.index1:
                assert self.index1_start >= 0, "Autorized values for index1_start : >= 0"
                assert self.index1_end >= self.index1_start, "Autorized values for index1_end : > index1_start"
            if self.index2:
                assert self.index2_start >= 0, "Autorized values for index2_start : >= 0"
                assert self.index2_end >= self.index2_start, "Autorized values for index2_end : > index2_start"
            if self.molecular1:
                assert self.molecular1_start >= 0, "Autorized values for molecular1_start : >= 0"
                assert self.molecular1_end >= self.molecular1_start, "Autorized values for molecular_end : > molecular1_start"
            if self.molecular2:
                assert self.molecular2_start >= 0, "Autorized values for molecular2_start : >= 0"
                assert self.molecular2_end >= self.molecular2_start, "Autorized values for molecular2_end : > molecular2_start"
                
            
        # Handle the possible exception by printing simple message for users.
        except ConfigParser.NoOptionError as E:
            print (E)
            print ("An option is missing in the configuration file")
            print ("Please report to the descriptions in the configuration file\n")
            exit()
        except ConfigParser.NoSectionError as E:
            print (E)
            print ("An section is missing in the configuration file")
            print ("Please report to the descriptions in the configuration file\n")
            exit()
        except (ValueError, AssertionError) as E:
            print (E)
            print ("One of the value in the configuration file is not in the correct format")
            print ("Please report to the descriptions in the configuration file\n")
            exit()
```
